<h3>Podsumowanie klasyfikacji<h3>


Najlepszą dokładnością wykazały się Regresja Logistyczna oraz Support Vector Machine - 84% trafonści


Dla Regresji logistycznej:

Precision: Spośród wszystkich opinii, które model zdiagnozował jako pozytywne, 91% z nich jest pozytywne w rzeczywistości. 76% z opinii zdiagnozowanych jako negatywne, faktycznie są opiniami negatywnymi. 40% z opinii zdiagnozowanych jako neutralne, faktycznie są opiniami neutralnymi.

Recall: Spośród wszystkich opinii pozytywnych, model poprwanie zdiagnozował 95% takich opinii. Spośród wszystkich opinii negatywnych, model poprwanie zdiagnozował 75% takich opinii. Spośród wszystkich opinii neutralnych , model poprwanie zdiagnozował zaledwie 29% takich opinii.

F1-score: Wartości Pozytywny = 0.93, Negatywny = 0.75, Neutralny = 0.33, mówią o tym jak dobrza algorytm klasyfikuje poszczególne przypadki. Algorytm w dość dobrym stopniu rozpoznaje opinie pozytywne, natomiast rozpoznawalność opinii neutralnych jest na bardzo niskim poziomie, jest to spowodowane liczebnością opinii w poszczególnych przyadkach.

Dla Support Vector Machine:

Precision: Spośród wszystkich opinii, które model zdiagnozował jako pozytywne, 85% z nich jest pozytywne w rzeczywistości. 80% z opinii zdiagnozowanych jako negatywne, faktycznie są opiniami negatywnymi. 60% z opinii zdiagnozowanych jako neutralne, faktycznie są opiniami neutralnymi.

Recall: Spośród wszystkich opinii pozytywnych, model poprwanie zdiagnozował 99% takich opinii. Spośród wszystkich opinii negatywnych, model poprwanie zdiagnozował 70% takich opinii. Spośród wszystkich opinii neutralnych , model poprwanie zdiagnozował zaledwie 3% takich opinii.

F1-score: Wartości Pozytywny = 0.91, Negatywny = 0.75, Neutralny = 0.03, mówią o tym jak dobrza algorytm klasyfikuje poszczególne przypadki. Algorytm w dość dobrym stopniu rozpoznaje opinie pozytywne, natomiast rozpoznawalność opinii neutralnych jest na bardzo niskim poziomie, jest to spowodowane liczebnością opinii w poszczególnych przyadkach.
